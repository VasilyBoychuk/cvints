&emsp;Для количественной оценки эффективности обнаружения алгоритмов обнаружения человека используется ряд способов. Поскольку проблема обнаружения человека может быть сформулирована в терминах бинарной классификации, в которой области-кандидаты являются либо содержащими человека, либо не включающими в себя человека, точность классификации/распознавания может рассматриваться как показатель эффективности обнаружения. В некоторых наборах данных (например, INRIA, CVC и DC) представлены изображения с людьми (положительные выборки) и не содержащие людей (отрицательные выборки), оценка может быть выполнена просто как тестирование системы распознавания объектов. В этом случае производительность обнаружения может быть представлена receiver operator characteristic (ROC), которая представляет собой компромисс между показателем истинно-положительных случаев (<b>True Positive Rates (TPR) </b>) и ложно-положительных (<b>False Positive Rates (FPR) </b>)

$\begin{equation*}
TPR = \frac{\text{True recognitions}}{\text{Positives samples}}
\end{equation*}$
$\begin{equation*}
FPR = \frac{\text{False recognitions}}{\text{Negatives samples}}
\end{equation*}$
&emsp;$\begin{equation*}
\text{(1)}
\end{equation*}$

&emsp;Обратите внимание, что в некоторых методах [15] TPR заменяется <b> Miss Rate (MR) </b>, определяемой как <b> MR = 1.0 - TPR </b>. FPR также можно рассматривать как ложноположительную частоту на одно окно (FPPW), поскольку каждая выборка представлена окном обнаружения.

&emsp;Хотя ROC с TPR/MR и FPR/FPPW часто используются для оценки алгоритмов детекции объектов (object detection algorithms), эта метрика имеет недостатки при сравнении методов обнаружения на наборах данных, которые явно не предоставляют образцы (т. е. обрезанные изображения). Как утверждают Agarwal&Roth [154], количество отрицательных образцов, генерируемых сканированием входного изображения с использованием окна обнаружения, зависит от многих факторов, таких как размер окна обнаружения, количество шкал, шаг шкалы и пространственный шаг. Эти параметры могут быть настроены по-разному в разных системах обнаружения. Таким образом, когда примеров отсутствия людей нет
 precision-recall оценка (PR) должна использоваться в качестве метрики для оценки и сравнения алгоритмов обнаружения человека. Определение отзыва и точности дается следующим образом:

&emsp;$\begin{equation*}
Recall = \frac{\text{True positives}}{\text{Annotated humans}}
\end{equation*}$
&emsp;$\begin{equation*}
Precision = \frac{\text{True positives}}{\text{True positives + False positives}}
\end{equation*}$
&emsp;$\begin{equation*}
\text{(2)}
\end{equation*}$

&emsp;Результат обнаружения (bounding box) считается истинно-положительным (TP), если он соответствует ground-truth разметке и ложно-положительным (FP или false alarm) в противном случае. Как можно заметить, метрика полнота (recall) соответствует TPR, в то время как точность (precision) не зависит от того, сколько окон обнаружения отбирается из тестового изображения.

&emsp;В [152, 9] спрогнозированный bounding box ($BB_{dt}$) считается совпавшим с аннотироваанной разметкой ($BB_{gt}$), если выполняется следующее неравенство:

&emsp;$\begin{equation*}
λ <  \frac{area(BB_{gt} ⋂ BB_{dt})}{area(BB_{gt} ⋃ BB_{dt})}, (3)
\end{equation*}$

где λ - заданное пользователем значение, представляющее степень перекрытия (наложения) спрогнозированного bounding box'а ($BB_{dt}$) на истинный результ ($BB_{gt}$). Это значение часто устанавливается равным 0,5 при оценке и сравнении детекторов человека [9,152].

&emsp;Аналогично PR-мере, Dolmar et al. [9] предложили использовать <b> false positive per image (FPPI) </b> вместо <b> FPPW vs MR </b> для оценки общего качества обнаружения. В оценке FPPI соседние окна обнаружения, которые классифицируются как содержащие людей, объединяются на этапе постобработки.  Non-maximal suppression (NMS)является наиболее широко используемым алгоритмом постобработки. В основном NMC проводится с использованием среднего сдвига или попарного максимального подавления [9].

&emsp;Затем производительность обнаружения оценивается на основе результатов объединенного обнаружения. В работе [9] показано, что существует расхождение между оценкой FPPW и FPPI, т. е. если один метод лучше другого при использовании FPPW, то он не может превзойти другой метод при использовании FPPI. Рисунок 20 (а) иллюстрирует расхождение в сравнении нескольких методов обнаружения человека (например, методов “Shapelet” и “HIKSvm”) в наборе данных INRIA с использованием оценки FPPW и FPPI. Более того, в работе [9], Dollár et al. утверждалось, что окна обнаружения, генерируемые оценкой FPPW, не совпадают с окнами обнаружения, генерируемыми во время оценки FPPI, и что оценка FPPW не охватывает все случаи, рассматриваемые оценкой FPPI. Рисунок 20 (b) иллюстрирует случаи, которые часто встречаются при оценке FPPI, но не включаются в оценку FPPW. Например, люди в положительных образцах при оценке FPPW должны быть хорошо центрированы в окнах обнаружения. Следовательно, ложных срабатываний, вызванных масштабированием и переводом людей и обнаружением частей тела не произойдет.

&emsp;Однако мы обнаружили, что описанная выше ситуация (т. е. несогласованность окон обнаружения, генерируемых в процессе оценки FPPW и FPPI) происходит из-за способа построения положительных и отрицательных выборок. Предположим, что нам дано изображение с аннотированными объектами (т. е. истинными объектами), мы сканируем это изображение с помощью окна обнаружения в различных масштабах и местоположениях. Затем каждое окно помечается как положительное или отрицательное окно на основе некоторого критерия, например представленного в формуле (3). Этот процесс приведет к наборам положительных и отрицательных выборок (windows). Обратите внимание, что на основе критерия в (3) соотношение положительного окна и содержащегося в нем человеческого объекта может изменяться, и человеческий объект также не может быть центрирован в окне. Кроме того, отрицательные окна не только представляют фон, но и захватывают части человеческих объектов. Мы проиллюстрируем этот процесс на рис. 21.

При использовании этой методологии положительные и отрицательные наборы включали бы все случаи, упомянутые на рис. 20 (b), то есть комбинация положительных и отрицательных множеств будет такой же, как и набор окон, генерируемых оценкой FPPI. Поэтому мы считаем, что оценка FPPW не является ошибочной в оценке эффективности обнаружения. Скорее, он просто стоит с другой точки зрения, точки зрения распознавания объектов. Кроме того, в некоторых приложениях, например для поиска изображений, оценка с использованием FPPW является разумной и, вероятно, предпочтительной, чем FPPI.

Оценка FPPW также имеет преимущества по сравнению с оценкой FPPI. Во-первых, FPPW можно использовать для сравнения признаков, человеческих дескрипторов и классификаторов. Действительно, если все алгоритмы обнаружения реализуются с использованием одной и той же экспериментальной установки (т. е. одного и того же количества шкал, одного и того же пространственного и масштабного шага), то будут созданы одни и те же положительные и отрицательные выборки, и FPPW будет валиден для оценки и сравнения. Во-вторых, оценка FPPW не зависит от NMS, в то время как расхождение между FPPW и FPPI может быть связано с NMS. Мы проиллюстрируем влияние NMS (в двух случаях: одна и та же NMS и разные NMS) на производительность обнаружения на рис. 22. Рисунок 22 (а) демонстрирует случай, в котором используется тот же самый NMS. На этом рисунке синие прямоугольники (пунктирной линией) представляют аннотированные объекты, зеленые и красные прямоугольники - истинные и ложные результаты обнаружения соответственно. Критерий, используемый для определения истинных и ложных обнаружений, представлен в (3). Рисунок 22 (a) объясняет, почему существует расхождение между оценкой FPPW и оценкой FPPI при сравнении методов обнаружения. Как показано на рис. 22 (а), алгоритм 2 имеет лучший индикатор FPPW, тогда как алгоритм 1 имеет лучший индикатор FPPI. Рисунок 22 (b) иллюстрирует влияние различных методов НМС на результаты обнаружения. Как показано, эти два метода NMS могут давать различные результаты обнаружения.

&emsp;В заключение следует отметить, что для точного и справедливого сравнения методы обнаружения человека должны оцениваться как с помощью FPPW, так и с помощью измерения FPPI. Кроме того, для оценки FPPW положительные и отрицательные окна должны быть сгенерированы таким образом, чтобы они совпадали с окнами, полученными в процессе оценки FPPI.
